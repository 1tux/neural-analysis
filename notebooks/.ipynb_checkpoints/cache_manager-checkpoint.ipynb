```python
import shelve
from pprint import pprint
#import sys
#sys.path.append("..")
import models

f = shelve.open("../cache/submodels")
keys = list(f.keys())
v = 10 ** 10
best_k = -1
best_edof = -1
aics = []
n_features = []
for k in keys:
    aic = f[k].gam_model.statistics_['AICc']
    edof = f[k].gam_model.statistics_['edof']
    aics.append(aic)
    n_features.append(k.count('F_X')+k.count('F_HD'))
    if aic < v:
        v = aic
        best_k = k
        best_edof = edof
    # print(k, aic, edof)

print(best_k, v, edof)
import matplotlib.pyplot as plt
#aics = [x-min(aics) for x in aics]
aics = [x+random.random() for x in aics]
import seaborn
seaborn.violinplot(x=n_features, y=aics, alpha=0.5, inner='stick')
plt.xlabel('n_of_features')
plt.ylabel('AIC')
```
